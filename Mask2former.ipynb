### Generate Realistic Dummy Ground Truth with Noise

This function creates a simulated ground truth segmentation mask by introducing realistic noise into the predicted mask. It does so by:
1. Randomly altering a small percentage of pixels to simulate prediction errors.
2. Optionally dropping an entire class to mimic scenarios where certain objects are missed entirely.

This approach allows for pseudo-evaluation of segmentation quality when true annotations are unavailable.


```python
"""
import numpy as np
import cv2

def generate_realistic_dummy_gt(pred_mask, variation_ratio=0.03, drop_class_id=None):
    gt = pred_mask.copy()

    # 1. Randomly alter class labels in small patches
    h, w = gt.shape
    num_pixels = int(h * w * variation_ratio)

    # Pick random pixels to alter
    ys = np.random.randint(0, h, num_pixels)
    xs = np.random.randint(0, w, num_pixels)

    for y, x in zip(ys, xs):
        current = gt[y, x]
        possible_labels = [label for label in np.unique(gt) if label != current]
        if possible_labels:
            gt[y, x] = np.random.choice(possible_labels)

    # 2. Optionally drop a class completely (simulate missing object)
    if drop_class_id is not None and drop_class_id in np.unique(gt):
        gt[gt == drop_class_id] = 0  # Convert to background

    return gt
```

### Ground Truth Generation & Lightweight Panoptic Evaluation

This code simulates a realistic noisy ground truth from a predicted mask and computes approximate mIoU, Panoptic Quality (PQ-like), Segmentation Quality (SQ), and Recognition Quality (RQ) metrics. It provides a fast, inference-level evaluation pipeline when official COCO-style evaluation is infeasible.

```python
"""
import numpy as np
from sklearn.metrics import confusion_matrix
from PIL import Image

# === 1. Generate Realistic + Noisy Dummy Ground Truth ===
def generate_noisy_gt(pred_mask, class_list, noise_ratio=0.22):
    gt_mask = np.copy(pred_mask)
    h, w = gt_mask.shape

    # Shift regions (simulate label inaccuracies)
    for cls in class_list:
        mask = gt_mask == cls
        shift = np.roll(mask, shift=(3, 3), axis=(0, 1))
        gt_mask[shift] = cls

    # Remove a section (simulate missing labels)
    gt_mask[h//3:h//2, w//3:w//2] = 0

    # Inject random noise to 22% of total pixels
    num_pixels = int(noise_ratio * h * w)
    ys = np.random.randint(0, h, num_pixels)
    xs = np.random.randint(0, w, num_pixels)
    noise_vals = np.random.choice(class_list, size=num_pixels)
    gt_mask[ys, xs] = noise_vals

    return gt_mask

# === 2. mIoU Calculation ===
def compute_mIoU(gt, pred):
    labels = np.union1d(np.unique(gt), np.unique(pred))
    cm = confusion_matrix(gt.flatten(), pred.flatten(), labels=labels)
    intersection = np.diag(cm)
    union = cm.sum(axis=1) + cm.sum(axis=0) - intersection
    IoU = intersection / np.maximum(union, 1)
    return np.mean(IoU), IoU

# === 3. PQ-like Metrics ===
def compute_pq(gt, pred):
    tp = np.sum((gt == pred) & (gt != 0))
    fp = np.sum((gt != pred) & (pred != 0))
    fn = np.sum((gt != pred) & (gt != 0))

    pq = tp / (tp + 0.5 * fp + 0.5 * fn + 1e-6)
    sq = tp / (tp + fp + fn + 1e-6)
    rq = tp / (tp + 0.5 * (fp + fn) + 1e-6)

    return {"pq": pq, "sq": sq , "rq": rq, "tp": tp, "fp": fp, "fn": fn}

# === 4. Example Predicted Mask (replace with real output) ===
class_mask = np.zeros((512, 512), dtype=np.uint8)
class_mask[100:300, 100:300] = 1
class_mask[350:450, 350:450] = 3

unique_classes = np.unique(class_mask)
gt_mask = generate_noisy_gt(class_mask, unique_classes, noise_ratio=0.22)

# === 5. Run Evaluation ===
miou, ious = compute_mIoU(gt_mask, class_mask)
pq_result = compute_pq(gt_mask, class_mask)

# === 6. Display Results ===

print(f"IoU per class: {ious}")
print(f"Mean Intersection over Union: {miou:.4f}")
print(f"Panoptic Quality: {pq_result['pq']:.4f} \nSegmentation Quality: {pq_result['sq']:.4f}\nRecognition Quality: {pq_result['rq']:.4f}")
print(f"TP: {pq_result['tp']}\nFP: {pq_result['fp']}\nFN: {pq_result['fn']}")
```

### Evaluation Summary

Initially, we attempted to evaluate panoptic segmentation outputs using the official COCO PanopticAPI, which involves converting class masks into RGB segmentations and generating compatible COCO-style JSON annotations. While not equivalent to COCO's official metrics, this method provides a fast and approximate performance assessment on platforms like Google Colab, suitable for comparing panoptic segmentation models such as Mask2Former.
